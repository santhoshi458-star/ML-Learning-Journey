```python
import os
from dotenv import load_dotenv

load_dotenv(override=True)
api_key=os.getenv('OPENAI_API_KEY')

if not api_key:
    print("No API key found")
elif not api_key.startswith("sk-proj-"):
    print("invalid API key")
else:
    print("API key found and looks good so far")
```

```python
#traditional REST API to hit Open AI model
import requests
headers={"Authorization":f"Bearer {api_key}","Content-type":"application/json"}
payload={"model":"gpt-5-nano",
         "messages":[{"role":"user","content":"Tell me a fun fact"}]
        }
```

```python
response=requests.post("https://api.openai.com/v1/chat/completions",headers=headers,json=payload)
```

```python
response.json()
```

```python
response.json()["choices"][0]["message"]["content"]
```

```python
#later with Chatcompletions API
from openai import OpenAI
client=OpenAI()
response=client.chat.completions.create(model="gpt-5-nano",messages=[{"role":"user","content":"tell me a fun fact"}])
print(response.choices[0].message.content)

```

```python
#run ollama model with chat completions api
OLLAMA_BASE_URL = "http://localhost:11434/v1"
ollama=OpenAI(base_url=OLLAMA_BASE_URL,api_key="ollama")
response = ollama.chat.completions.create(model="llama3.2:1b", messages=[{"role": "user", "content": "Tell me a fun fact"}])

print(response.choices[0].message.content)
```
