```python
df=pd.read_csv('/kaggle/input/sales-data/sales_data_visualization.csv')
```

```python
df.head()
df.info()
```

```python
#Histogram for sales
import matplotlib.pyplot as plt
plt.figure(figsize=(10,5))
plt.hist(df['Sales'],bins=15,color='green',edgecolor='black')
plt.title("Histogram for Sales")
plt.xlabel("Frequency")
plt.ylabel("Sales")
plt.show()
```

```python
#Box plot for Profit
import seaborn as sns
plt.figure(figsize=(8,6))
sns.boxplot(df['Profit'],color='lightgreen')
plt.title('Box plot for Profit')
plt.show()
```

```python
#Line plot for time series sales trend
plt.figure(figsize=(10,8))
plt.plot(df['Date'],df['Sales'],marker='o',linestyle='-',color='purple')
plt.xlabel("Date")
plt.ylabel("Sales")
plt.xticks(rotation=45)
plt.title("Daily sales trend")
plt.show()
```

```python
#Bivariate Analysis
#Scatter plot sales vs advertising spend
plt.figure(figsize=(10,8))
plt.scatter(df['Advertising_Spend'],df['Sales'],color='orange',alpha=1)
plt.xlabel('Advertising_Spend')
plt.ylabel("Sales")
plt.title('Advertising_Spend vs Sales')
plt.show()
```

```python
#Bar plot Region vs Avg Sales
plt.figure(figsize=(10,8))
sns.barplot(x='Region',y='Sales',data=df,estimator='mean')
plt.title("Region vs Average Sales")
plt.show()
```
